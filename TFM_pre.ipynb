```python
# -*- coding: utf-8 -*-

#Imports
import os
import pandas as pd
import csv
import json
from pandas.io.json import json_normalize
import numpy as np
import re
from datetime import datetime
import datetime
import time
import glob
import io
```

```python
input_data = os.listdir('../TFM/json_data_sorted/2015/diciembre2015/')

count = 0
list_names = []
for file_name in input_data:
    if file_name.endswith(".json"):
        count = count +1
        list_names.append(file_name)
    
#print(list_names)
print('Total Elementos: ',count)
```

```python
#Definiendo fecha y hora de cada archivo
p_1 = '\d{2}-\d{2}-\d{4}'
p_2 = '\d{2}-\d{2}-\d{2}\.'

d_1 =[] 
t_1=[]
for x_2 in list_names:
    d_2 = re.search(p_1,x_2).group()
    d_1.append(d_2)
    
    t_2 = re.search(p_2, x_2).group()
    t_2 = re.sub(r'-',':',t_2)
    t_2 = re.sub(r'\.','',t_2)
    t_1.append(t_2)
    
print('Fecha: ',d_1)
print('Hora: ',t_1)
```

```python
print('Fecha','      ','Hora')
print('-----','      ','----')
for x_3,x_4 in zip(d_1,t_1):
    print(x_3,' ',x_4)
```

```python
#Función para poder normalizar archivos json
def flatten_json(y):
    out = {}

    def flatten(x, name=''):
        if type(x) is dict:
            for a in x:
                flatten(x[a], name + a + '_')
        elif type(x) is list:
            i = 0
            for a in x:
                flatten(a, name + str(i) + '_')
                i += 1
        else:
            out[name[:-1]] = x

    flatten(y)
    return out
```

```python
#Definiendo fecha y hora de cada archivo
pat_1 = '\d{2}-\d{2}-\d{4}'
pat_2 = '\d{2}-\d{2}-\d{2}\.'
j_files = glob.glob("../TFM/json_data_sorted/2015/diciembre2015/*.json")

#Construyendo nuevo objeto json añadiendo la fecha y la hora correspondiente
date =[] 
time=[]
q=[]

    
for filename in j_files:
    date = re.search(pat_1,filename).group()
    time = re.search(pat_2, filename).group()
    time = re.sub(r'-',':',time)
    time = re.sub(r'\.','',time)
    with open(filename,'r') as file_1:
        try:
            data = json.loads(file_1.read())
        except json.JSONDecodeError as e:
            print('Error Parseo:',e)
            pass
        for item in data['features']:
            item['properties']['fecha'] = date
            item['properties']['hora'] = time
            #json_dates = json.dumps(item)
            q.append(item)
```

```python
for z in q[:1]:
    print(json.dumps(z,indent=3))
```

```python
import time
start = time.process_time()
#Normalizamos el formato json para transformarlo en un dataframe
q1=[]
for row in q:
    flat = flatten_json(row)
    n_json = json_normalize(flat)
    q1.append(n_json)

    
    
end = time.process_time()
resultado = end - start

m, s = divmod(resultado, 60)
h, m = divmod(m, 60)
print ("%02d:%02d:%02d" % (h, m, s))
```

```python
start_2 = time.process_time()
#Convertimos los datos a Pandas Dataframe
df_json = pd.concat([pd.DataFrame(x) for x in q1],sort=True,ignore_index=True)

end_2 = time.process_time()
resultado_2 = end_2 - start_2
print(resultado_2)
```

```python
#Conversion segundos a horas
m_2, s_2 = divmod(resultado_2, 60)
h_2, m_2 = divmod(m_2, 60)
print ("%02d:%02d:%02d" % (h_2, m_2, s_2))
```
